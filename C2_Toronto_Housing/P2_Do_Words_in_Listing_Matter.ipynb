```python
from sklearn.feature_extraction.text import CountVectorizer
word_vectorizer = CountVectorizer(ngram_range=(4,4), analyzer='word')
sparse_matrix = word_vectorizer.fit_transform(df['name'])
frequencies = sum(sparse_matrix).toarray()[0]


freq_data = pd.DataFrame(frequencies, index=word_vectorizer.get_feature_names_out(), columns=['frequency'])
print(len(freq_data))
```

```python
df1 = freq_data.sort_values(by=['frequency'], ascending=False)
df1.head(30)
```

```python
# Select the first five rows from the data set
td = pd.DataFrame(sparse_matrix.todense()).iloc[:10]
td.columns = word_vectorizer.get_feature_names_out()
term_document_matrix = td.T
term_document_matrix.columns = ['Doc '+str(i) for i in range(0, 10)]
term_document_matrix['total_count'] = term_document_matrix.sum(axis=1)
term_document_matrix.drop(columns=['total_count'],inplace=True)
term_document_matrix_new =term_document_matrix.T
print(term_document_matrix_new)

```
